```python

# notebooks/example_notebook.ipynb

import sys
import os
sys.path.append(os.path.abspath('../scripts'))
```

```python
from database import connect_to_db
from user_engagement import get_engagement_metrics, normalize_metrics, cluster_users
from analysis import bivariate_analysis
```

```python
# Connect to the database
engine = connect_to_db()
```

```python

# Fetch engagement metrics
engagement_metrics = get_engagement_metrics(engine)
```

```python
# Normalize metrics
columns_to_normalize = ['session_frequency', 'total_duration', 'total_traffic']
engagement_metrics = normalize_metrics(engagement_metrics, columns_to_normalize)
```

```python
# Perform clustering
engagement_metrics, kmeans = cluster_users(engagement_metrics, [f"{col}_norm" for col in columns_to_normalize])
```

```python
# Bivariate analysis
bivariate_analysis(engagement_metrics, x="total_duration_norm", y="total_traffic_norm", kind="scatter", title="Cluster Analysis: Duration vs Traffic", xlabel="Normalized Duration", ylabel="Normalized Traffic")
```
